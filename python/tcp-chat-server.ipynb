```python
import socket
import threading
```

```python
HOST = "192.168.219.154"
PORT = 7070
```

```python
def messanger_thread(socketFrom, socketTo):
    """ Receive one message from socketFrom send it back to socketTo, 
    then close socket """
    try:
        connFrom, addrFrom = socketFrom
        connTo, addrTo = socketTo
        netIn = connFrom.makefile(mode='rb', encoding='utf-8')
        netOut = connTo.makefile(mode='wb', encoding='utf-8')
        while True:
            msg = netIn.readline()
            if not msg :
                break

            print(msg)
            netOut.write(msg)
            netOut.flush()
    except (ConnectionError, BrokenPipeError):
        print('Socket error')
    finally:
        print('Closed connection to {}'.format(addrFrom))
        connFrom.close()
        print('Closed connection to {}'.format(addrTo))
        connTo.close()
```

```python
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
sock.bind((HOST, PORT))
sock.listen(10)
```

```python
print('Listening at', sock.getsockname())

while True:
    try:
        print ("Wait for Two Client Connection")
        conn1, addr1 = sock.accept()
        conn2, addr2 = sock.accept()
        print('We have accepted a connection ', addr1)
        print(' Socket name:', conn1.getsockname())
        print(' Socket peer:', conn1.getpeername())
        print('We have accepted a connection ', addr2)
        print(' Socket name:', conn2.getsockname())
        print(' Socket peer:', conn2.getpeername())
    
        thread1 = threading.Thread(target=messanger_thread, args=[(conn1, addr1), (conn2, addr2)], daemon=True)
        thread1.start()
        
        thread2 = threading.Thread(target=messanger_thread, args=[(conn2, addr2), (conn1, addr1)], daemon=True)
        thread2.start()
        
    except KeyboardInterrupt:
        print("Exited by user")
        break
```
